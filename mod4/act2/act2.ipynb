```python
from ordered_set import OrderedSet
import pandas as pd
import numpy as np
from numpy.typing import NDArray
```

```python
def vectorizeWords(s: str, t: str) -> tuple[NDArray[np.int64], NDArray[np.int64]]:
    def clean_text(text: str) -> list[str]:
        return ''.join(c.lower() if c.isalnum() or c == " " else '' for c in text).split()

    s_words = clean_text(str(s))
    t_words = clean_text(str(t))

    words = OrderedSet(s_words + t_words)

    map_s = {}
    map_t = {}

    for word in s_words:
        if word in map_s.keys():
            map_s[word] += 1
        else:
            map_s[word] = 1
            
    for word in t_words:
        if word in map_t.keys():
            map_t[word] += 1
        else:
            map_t[word] = 1

    vec_s = []
    vec_t = []

    for word in words:
        if word in map_s.keys():
            vec_s.append(map_s[word])
        else:
            vec_s.append(0)

        if word in map_t.keys():
            vec_t.append(map_t[word])
        else:
            vec_t.append(0)

    return np.array(vec_s), np.array(vec_t)
```

```python
def calculateSimilarity(vec1: NDArray[np.int64], vec2: NDArray[np.int64]) -> float:
    dot = np.dot(vec1, vec2)
    vec1_mag = np.linalg.norm(vec1)
    vec2_mag = np.linalg.norm(vec2)

    return dot / (vec1_mag * vec2_mag)
```

```python
data = pd.read_csv("quora.csv")
data.dropna(inplace=True)

result = pd.DataFrame(columns=["cosine_distance", "q1_vector", "q2_vector"])

for index, row in data.iterrows():
    q1, q2 = row["question1"], row["question2"]
    v1, v2 = vectorizeWords(q1, q2)
    sim = calculateSimilarity(v1, v2)

    result.loc[index] = [sim, v1, v2]

data["cosine_distance"] = result["cosine_distance"]
data["q1_vector"] = result ["q1_vector"]
data["q2_vector"] = result ["q2_vector"]

data.to_csv("similarity.csv")
data.head()

```
