Lanzar un dado 1000 veces y mostrar sus resultados

```python
from random import randint
import matplotlib.pyplot as plt

results = []
throws = 1000

for i in range(throws):
    results.append(randint(1, 6))

frequencies = [results.count(i) for i in range(1, 7)]
```

```python
fig, ax = plt.subplots()

ax.bar([i for i in range(1,7)], frequencies)
ax.set_title(f"Resultados de lanzar el dado {throws} veces")
ax.set_xlabel("N√∫mero de la cara")
ax.set_ylabel("Frecuencia")
plt.show()
```

```python
print("El promedio de los resultados es: ", sum(results) / len(results))
```
